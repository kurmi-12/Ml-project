```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
placement_data = pd.read_csv('/content/placement-dataset.csv')
placement_data.head()
```

```python
X =placement_data.iloc[:,:-2].values
Y =placement_data.iloc[:,-1].values
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_test)
```

```python
print(Y_test)
```

```python
y_test= Y_test.reshape(len(Y_test),1)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train= sc.fit_transform(X_train)
X_test= sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state= 0)
classifier.fit(X_train,Y_train)
```

```python
y_predict= classifier.predict(X_test)
```

```python
print(y_predict)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm= confusion_matrix(Y_test, y_predict)
print(cm)
accuracy_score(Y_test, y_predict)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train,Y_train,clf=classifier, legend=2)
```

```python
plot_decision_regions(X_test,y_predict,clf=classifier, legend=2)
```

```python
plot_decision_regions(X_test,Y_test,clf=classifier, legend=2)
```
